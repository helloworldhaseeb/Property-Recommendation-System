```python
def Marla_to_Kanal(area_str):
    # Replace commas with nothing to handle cases like '1,000 marla'
    area_str = area_str.replace(',', '')
    
    if 'Marla' in area_str:
        Marla = float(area_str.split()[0])
        Kanal = Marla / 20.0  # 1 kanal = 20 marla
        return Kanal
    elif 'Kanal' in area_str:
        Kanal = float(area_str.split()[0])
        return Kanal
    else:
        return None  # Return None if neither 'Marla' nor 'Kanal' is found
# Assuming fprops['area'] contains strings like '500 Marla' or '10 Kanal'
fprops['area'] = fprops['area'].apply(lambda x: Marla_to_Kanal(x) if isinstance(x, str) else x)
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from scipy.spatial import distance

# Assuming fprops is your DataFrame with property data
df = fprops.copy()

# Check for and handle NaN or inf values
df = df.dropna()  # Drop rows with NaN values (if appropriate for your dataset)
df = df.replace([np.inf, -np.inf], np.nan).dropna()  # Drop rows with inf values

# Normalize numerical features (optional but recommended)
scaler = MinMaxScaler()
df[['latitude', 'longitude', 'baths', 'bedrooms', 'area', 'standardized_price']] = scaler.fit_transform(df[['latitude', 'longitude', 'baths', 'bedrooms', 'area', 'standardized_price']])

# Function to calculate Euclidean distance between two properties
def calculate_similarity(property1, property2):
    features = ['latitude', 'longitude', 'baths', 'bedrooms', 'area', 'standardized_price']
    dist = distance.euclidean(property1[features], property2[features])
    return dist

# Function to recommend similar properties based on a given property
def get_similar_properties(property_id, df, num_recommendations=5):
    property_to_compare = df[df['property_id'] == property_id].squeeze()
    similarities = []

    for idx, row in df.iterrows():
        if row['property_id'] != property_id:
            dist = calculate_similarity(property_to_compare, row)
            similarities.append((row['property_id'], dist))

    similarities.sort(key=lambda x: x[1])  # Sort by distance (lower is more similar)

    recommendations = similarities[:num_recommendations]
    return recommendations

# Example usage
property_id = 40538586  # Example property ID
similar_properties = get_similar_properties(property_id, df)

# Convert similarity distances to percentages
num_features = 6
max_distance = np.sqrt(num_features)

similar_property_ids = [prop_id for prop_id, dist in similar_properties]
similarity_distances = [dist for prop_id, dist in similar_properties]
similarity_percentages = [(1 - (dist / max_distance)) * 100 for dist in similarity_distances]

# Print similarity distances and percentages
print(f"Properties similar to property {property_id}:")
for prop_id, dist, percent in zip(similar_property_ids, similarity_distances, similarity_percentages):
    print(f"Property {int(prop_id)}, Similarity Distance: {dist:.6f}, Similarity Percentage: {percent:.2f}%")
```
